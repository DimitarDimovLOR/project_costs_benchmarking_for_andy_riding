```python
subcontractorsData = pd.read_sql_query("""SELECT DISTINCT a.avm_num, b.avm_name

FROM [DW_Mirrored_Data].[src_coins_uk].[ap_invoice_PAR1] as a

jOIN [DW_Mirrored_Data].[src_coins_uk].[ap_vendor] as b on a.avm_num = b.avm_num


WHERE ain_anal like 'V3588'
OR ain_anal like 'PH1012'
OR ain_anal like 'TM065'
  OR ain_anal like 'V3905'
  OR ain_anal like 'V7706'
  OR ain_anal like 'V7710'
   OR ain_anal like 'V3900'
    OR ain_anal like 'W3871'
	 OR ain_anal like 'W3509'
	  OR ain_anal like 'W3573'
	   OR ain_anal like 'PRWAL306'
	    OR ain_anal like 'PRWAL316'
		 OR ain_anal like 'A3880'
		  OR ain_anal like 'HN251'
		   OR ain_anal like 'Y3902'
		    OR ain_anal like 'CHM3871'
			 OR ain_anal like 'EIP3871'
			  OR ain_anal like 'GRC3871'
			   OR ain_anal like 'SEL3871'
			    OR ain_anal like 'EPH3871'
ORDER BY a.avm_num ASC
""",conn)

subcontractorsDataHistory = pd.read_sql_query("""SELECT DISTINCT a.avm_num, b.avm_name

FROM [DW_Mirrored_Data].[src_coins_uk].[ap_invoice_PAR1_History] as a

jOIN [DW_Mirrored_Data].[src_coins_uk].[ap_vendor_History] as b on a.avm_num = b.avm_num


WHERE ain_anal like 'V3588'
OR ain_anal like 'PH1012'
OR ain_anal like 'TM065'
  OR ain_anal like 'V3905'
  OR ain_anal like 'V7706'
  OR ain_anal like 'V7710'
   OR ain_anal like 'V3900'
    OR ain_anal like 'W3871'
	 OR ain_anal like 'W3509'
	  OR ain_anal like 'W3573'
	   OR ain_anal like 'PRWAL306'
	    OR ain_anal like 'PRWAL316'
		 OR ain_anal like 'A3880'
		  OR ain_anal like 'HN251'
		   OR ain_anal like 'Y3902'
		    OR ain_anal like 'CHM3871'
			 OR ain_anal like 'EIP3871'
			  OR ain_anal like 'GRC3871'
			   OR ain_anal like 'SEL3871'
			    OR ain_anal like 'EPH3871'
ORDER BY a.avm_num ASC
""",conn)
```

```python
scccInvoiceData = scccInvoiceData.append(scccInvoiceDataHistory, sort = FALSE)
subcontractorsData = subcontractorsData.append(subcontractorsDataHistory, sort = FALSE)
```

```python
scccInvoiceData["CostCodes"] = scccInvoiceData.jcc_cc.str[-4:]
```

```python
mappingTable = pd.read_excel("CostCodes_mapping.xlsx")
```

```python
masterMapDesc = dict(mappingTable[["CODE","DESCRIPTION"]].values)
masterMapAlloc = dict(mappingTable[["CODE","ALLOCATION"]].values)

subContractorsMap = dict(subcontractorsData[["avm_num","avm_name"]].values)

scccInvoiceData["DESCRIPTION"] = scccInvoiceData.CostCodes.map(masterMapDesc)
scccInvoiceData["ALLOCATION"] = scccInvoiceData.CostCodes.map(masterMapAlloc)
scccInvoiceData["SUBCONTRACTOR"] = scccInvoiceData["avm_num"].map(subContractorsMap)
```
